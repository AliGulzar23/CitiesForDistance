```python
import numpy as np
import os
from math import radians, cos, sin, asin, sqrt,pi, atan2
import csv
import pandas as pd
fileName = "CityInfo"
file = open(fileName +".csv","r")
cityNames = []
lat = []
lon = []
for x in file:
    entity = x.split(',')
    cityNames.append( entity[0])
    lat.append(entity[1])
    lon.append(entity[2])
file.close
array = np.zeros(len(lat)**2)
distanceMatrix = array.reshape(len(lat),len(lat))
radius = 6371

for x in range(0,len(lat)):
    for y in range(0,len(lat)):
        
        phi1 = int(lat[x]) * pi/180
        phi2 = int(lat[y]) * pi/180
        dphi = (int(lat[y])-int(lat[x])) * pi/180
        dland = (int(lon[y])-int(lon[x])) * pi/180
        a = sin(dphi/2) **2 +cos(phi1) * cos(phi2)* (sin(dland/2) **2)
        c = 2* asin(sqrt(a))
        km = radius*c 
        distanceMatrix[x][y] = km

print(distanceMatrix[0])
#matrix = pd.DataFrame(distanceMatrix).dropna
#matrix.to_csv("matrix.csv")
```
